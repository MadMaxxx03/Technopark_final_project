```python
import torch
import pandas as pd
import numpy as np
import random
import kan
from IPython.display import clear_output
import matplotlib.pyplot as plt
from torch import nn
import sklearn
from tqdm import tqdm
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
'''
num - the number of grid intervals. Default: 5.
k - the order of piecewise polynomial. Default: 3.
'''

class myModel(nn.Module):
    def __init__(self, device):
        super().__init__()
        # self.l1 = kan.KANLayer(in_dim=3, out_dim=3, num=5, k=3,
        #                      device=device)
        
        self.l2 = kan.KANLayer(in_dim=1, out_dim=1, num=10, k=3,
                             device=device)
        
    def forward(self, x):
        # x = self.l1(x)[0]
        x = self.l2(x)[0]

        return x
        
    def get_params(self):
        return self.parameters()
```

```python
xs = np.linspace(1, 10, 5000)
# ys = np.linspace(2, 20, 500)
```

```python
fs = np.sin(xs)
```

```python
plt.plot(xs, fs)
```

```python
data = pd.DataFrame(np.vstack([xs, fs]).T)
```

```python
from torch.utils.data import Dataset, DataLoader

class myDataset(Dataset):
    def __init__(self, data_frame):
        self.data = data_frame

    def __len__(self):
        return self.data.shape[0]

    def __getitem__(self, indx):
        data = self.data.iloc[indx].values
        x = data[:-1]
        y = data[-1]
        return x, y
```

```python
train_df, test_df = sklearn.model_selection.train_test_split(data, train_size=0.5)
```

```python
train_ds = myDataset(train_df)
test_ds = myDataset(test_df)
```

```python
train_loader = DataLoader(train_ds, batch_size=8, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=8)
```

```python
def train(model, train_loader, test_loader, loss_fun, optimizer, scheduler=None, epoch=15, plot=True, device='cpu'):
    PATH = 'model1.pth'

    loader = {'train': train_loader, 'test': test_loader}
    losses = {'train': [], 'test': []}
    accuracy = {'train': [], 'test': []}
    best_test_loss = torch.inf

    for epoc in range(epoch):
        train_loss = 0
        test_loss = 0
        train_acc = 0
        test_acc = 0
        print(f"Epoch {epoc + 1}/{epoch}:")
        for phase in ['train', 'test']:
            print_step = int(len(loader[phase]) / 4)
            if phase == 'train':
                model.train()
            else:
                model.eval()

            for batch_idx, data in tqdm(enumerate(loader[phase])):
                x, y = data
                
                x = x.to(device)
                y = y.to(device)

                predict = model(x)

                loss = loss_fun(predict, y)

                if phase == 'train':
                    optimizer.zero_grad()

                    train_loss += loss.item()
                    loss.backward()

                    optimizer.step()

                    if batch_idx % print_step == 0:
                         print(f"{' ' * 6}Iter: {batch_idx + 1}/{len(loader[phase])} BatchLoss: {loss:.4f}")

                else:
                    test_loss += loss.item()

        losses['train'].append(train_loss / len(loader['train']))
        losses['test'].append(test_loss / len(loader['test']))

        # accuracy['train'].append(train_acc / len(loader['train']))
        # accuracy['test'].append(test_acc / len(loader['test']))
        
        clear_output(True)
        if plot:
            plt.figure(figsize=(12, 5))
            plt.subplot(1, 2, 1)
            plt.plot(losses['train'], label='Train')
            plt.plot(losses['test'], label='Test')
            plt.xlabel('Epochs', fontsize=16)
            plt.ylabel('Loss', fontsize=16)
            plt.legend(loc=0, fontsize=16)
            plt.grid('on')
            # plt.subplot(1, 2, 2)
            # plt.plot(accuracy['train'], label='Train')
            # plt.plot(accuracy['test'], label='Test')
            # plt.xlabel('Epochs', fontsize=16)
            # plt.ylabel('Accuracy', fontsize=16)
            # plt.legend(loc=0, fontsize=16)
            # plt.grid('on')
            plt.show()

        if losses['test'][-1] < best_test_loss:
            best_test_loss = losses['test'][-1]
            checkpoint = {
                'model': model.state_dict(),
                'optimizer': optimizer.state_dict(),
                'scheduler': scheduler.state_dict() if scheduler else 0,
                'losses': losses,
                'accuracy': accuracy
                }
            torch.save(checkpoint, PATH)

        print(f"TestLoss: {losses['test'][-1]:.4f}  BestLoss: {best_test_loss:.4f}")

        if scheduler:
            scheduler.step()
```

```python
model = myModel(device)
model = model.double().to(device)
loss_fun = nn.MSELoss()
optimizer = torch.optim.Adam(model.get_params(), lr=0.005)
```

```python
train(model, train_loader, test_loader, loss_fun, optimizer, epoch=20, plot=True, device=device)
```

```python
xs = torch.linspace(-10, 10, 1000)
model.eval()
preds = []
with torch.no_grad():
    for x in tqdm(xs):
        x = x.to(device)
        x = x.unsqueeze(dim=-1).unsqueeze(dim=-1)
        preds.append(model(x).numpy().item())
```
